```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2

# Define the two sets
set1 = {'B020010K11RIK', 'RIAN', 'GM20721', 'CELSR1', 'UBE2O', 'SRP54B', 'DTL', 'ZFP335', 'TOMM6', 'PYCRL',
        'H2AC14-PS', 'BCAT2', 'SF3B4', 'PER1', 'DERPC', 'GM26448', 'RNU1b6', 'GM26232', 'PTGS2', 'PADI3',
        'LDHB', 'ERO1A', 'MPO', 'PDCD6IP', 'LTF', 'MYH9', 'LDHC', 'TKT', 'VDAC2', 'EIF5A', 'PADI1', 'PADI4',
        'ERO1B', 'PADI2', 'LCP1', 'SLC25A31', 'HSPA9', 'ANXA2', 'CAT', 'EIF5A2', 'PADI6', 'SLC25A5', 'PADI3',
        'ENO1', 'SOD1', 'PTMA', 'ACTN4', 'PRKCA', 'TUBA1B', 'EEF2', 'VDAC3', 'ELANE', 'AK2', 'MDH2', 'CALU',
        'LYZ1', 'LYZ2', 'HSP90AB1', 'HSPA5', 'P4HB', 'CCT7'}

set2 = {'PADI3', 'LDHB', 'ERO1A', 'HNRN', 'MPO', 'PDCD6IP', 'LTF', 'MYH9', 'LDHC', 'TKT', 'VDAC2', 'EIF5A', 'PADI1',
        'PADI4', 'ERO1B', 'PADI2', 'LCP1', 'EIF5ALI', 'SLC25A31', 'HSPA9', 'ANXA2', 'CAT', 'EIF5A2','ATP5FA1', 'PADI6', 'SLC25A5',
        'PADI3', 'ENO1', 'SOD1', 'MNDA', 'PTMA', 'ACTN4', 'PRKCA', 'TUBA1B', 'AS100A9', 'EEF2', 'VDAC3', 'ELANE', 'A6NHL2', 'AK2', 'LDHAL6A', 'MDH2',
        'CALU', 'TUBA4B', 'LYZ', 'HSP90AB1','HIST2H3C', 'HSPA5', 'P4HB', 'CCT7'}

# Create the Venn diagram
venn_diagram = venn2([set1, set2], set_labels=('Set 1', 'Set 2'))

# Customize the diagram (optional)
venn_diagram.get_label_by_id('10').set_text('Set 1')
venn_diagram.get_label_by_id('01').set_text('Set 2')
venn_diagram.get_label_by_id('11').set_text('Common')

# Get the counts for each set and common genes
set1_count = len(set1)
set2_count = len(set2)
common_genes = set1.intersection(set2)
common_count = len(common_genes)

# Add annotations for counts inside the Venn diagram
venn_diagram.get_label_by_id('10').set_text('')
venn_diagram.get_label_by_id('01').set_text('')
venn_diagram.get_label_by_id('11').set_text('')
venn_diagram.get_label_by_id('11').set_text(f'Common: {common_count}')
venn_diagram.get_label_by_id('10').set_text(f'Set 1: {set1_count}')
venn_diagram.get_label_by_id('01').set_text(f'Set 2: {set2_count}')

# Set font properties for the labels
font_props = {'weight': 'bold'}
for label_id in ['10', '01', '11']:
    venn_diagram.get_label_by_id(label_id).set_fontproperties(font_props)

# Add annotations outside the Venn diagram with arrow lines
for text, position, color in [('Set 1', (-0.25, 0.45), 'green'), ('Set 2', (0.75, 0.45), 'orange'), ('Common', (0.4, -0.2), 'blue')]:
    plt.annotate(text, xy=position, xytext=(position[0], position[1] - 0.2), ha='center', fontsize=12, weight='bold',
                 arrowprops=dict(facecolor='black', arrowstyle='-|>'))

# Add common gene names to the legend
if common_genes:
    legend_text = ', '.join(common_genes)
    plt.legend([venn_diagram.get_patch_by_id('11')], [f'Common Genes: {legend_text}'], loc='best', prop={'weight': 'bold', 'size': 8})

# Display the diagram
plt.title('Venn Diagram - Set 1 and Set 2')
plt.show()
```
